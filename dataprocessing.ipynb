```python
import pandas as pd
```

```python
Disaster = pd.read_csv("DisasterDeclarationsSummaries.csv")
```

```python
Disaster.shape
Disaster.head()
```

```python
COVID = Disaster[(Disaster['declarationTitle'] == 'COVID-19') | (Disaster['declarationTitle'] == 'COVID-19 PANDEMIC')].reset_index()
```

```python
COVID['DeclarationDate'] = pd.to_datetime(COVID['declarationDate'].str[:10])
```

```python
COVID.head()
```

```python
N_declaration_by_date_and_state = COVID.groupby(['DeclarationDate', 'state']).agg(cases = ('index', 'count'))
```

```python
N_declaration_by_date_and_state.to_csv("N_declaration_by_date_and_state.csv")
```
