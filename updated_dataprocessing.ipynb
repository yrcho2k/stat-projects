```python
import pandas as pd
```

```python
Disaster = pd.read_csv("DisasterDeclarationsSummaries.csv")
```

```python
Disaster.shape
Disaster.head()
```

```python
COVID = Disaster[(Disaster['declarationTitle'] == 'COVID-19') | (Disaster['declarationTitle'] == 'COVID-19 PANDEMIC')].reset_index()
```

```python
COVID['DeclarationDate'] = pd.to_datetime(COVID['declarationDate'].str[:10])
```

```python
COVID.head()
```

```python
N_declaration_by_date_and_state = COVID.groupby(['DeclarationDate', 'state']).agg(cases = ('index', 'count')).reset_index()
```

```python
# data source: https://www.kaggle.com/datasets/tennerimaheshwar/us-state-and-territory-latitude-and-longitude-data?resource=download
Geo = pd.read_csv("US_GeoCode.csv")
```

```python
Geo.columns = ["state", "latitude", 'longitude', 'name']
```

```python
Geo.head()
```

```python
# merge data sets
Output_df = N_declaration_by_date_and_state.merge(Geo, on = 'state', how = 'left')
```

```python
Output_df.head(5)
```

```python
Output_df.to_csv("cleaned_data.csv")
```
